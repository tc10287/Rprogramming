Identifying and Handling Missing Data
‘NA’ is a special symbol that represents a missing value. Typically, when data is read from an external source, the missing values are replaced by NA. We can also initialise a vector using the symbol ‘NA’. Missing values are handled using the functions ‘is.na()’ and ‘anyNA().’

```r
employee_age=c(24,45,NA,43,28)
is.na(employee_age) #if the particular elememt is NA
anyNA(employee_age) #if any NA is there
```

Special Symbols
‘Inf’ and ‘NaN’ are special symbols used to represent infinity and ‘not a number’ or unknown in R, respectively. These symbols are of special importance in computations; such values will have to be handled in a systematic manner.

```r
1/0 #infinity
2/0
1000/0
0/0 #not a number
```

**MATRIX**

A matrix data structure is a tabular arrangement of objects of the same data type. It is like an atomic vector or a vector, but it can be of one or two dimensions. Therefore, it has a rectangular arrangement of numbers.

Creation of Matrix

A Matrix object can be created using the function ‘matrix().’ It takes the following arguments:
1. Elements of the matrix as a single vector
2. Number of rows
3. Number of columns

By default, the elements are arranged by columns. If the elements need to be arranged by rows, then the argument ‘byrow’ needs to be set to TRUE. By default, the rows and columns are given the numbers 1, 2, etc. However, we can specify the names for the rows and the columns for better identification. This is done using the function ‘rownames()’ and ‘colnames()’.


```r
#syntax
#matrix_name=matrix(vector,nrow=3,ncol=8)
matrix1=matrix(c(2,3,4,5,6,7),nrow=3,ncol=2)
print(matrix1)
matrix2=matrix(c(96,81,22,52,64,73),nrow=3,ncol=2,byrow = TRUE) #arrange the elements by row
print(matrix2)
rownames(matrix2)=c('dolma','john','laney')
colnames(matrix2)=c('math','physics')
print(matrix2)
```

Accessing Elements of a Matrix

The elements in the matrix are accessed using the indexes. Two indexes are needed to specify the row index and the column index. This should be specified within a square bracket and delimited using a comma.

```r
matrix2[2,1] #value on 2nd row, 1st column
matrix1[2,2]
```

**Data Frame**

A data frame is a list of lists with each sublist of the same length. It is equivalent to a rectangular list. A data frame is typically used to store data that are read from text/CSV files by retaining the underlying structure such as row names, column names, etc. A data frame can also be created manually

Creation of Data Frame

The function to create a dataframe is ‘data.frame()’. It takes as input various lists of equal length. Each of this list will correspond to a column in the dataframe. We can assign the names to the columns and rows using the functions ‘rownames()’ and ‘colnames().’

```r
#dataframes
student_name=c('ten','hazel','tara','melon')
student_ID=c(1012,1034,1035,3456)
student_degree=c('Bsc','Msc','Phd','Bsc')
student_marks=c(96.5,89,78.7,92.5)
student_data=data.frame(student_name,student_ID,student_degree,student_marks)
print(student_data)
```

Head and Tail Functions

head prints first 5 rows

tail prints last 5 rows

```r
head(student_data,2)
tail(student_data,1)
```

```r
dim(student_data) #dimensions of the data frame
nrow(student_data) #number of rows of the dataframe
ncol(student_data) #number of cols of the dataframe
```

**Accessing Elements of a Data Frame/Accessing Elements of a Column**

Accessing the elements of a slot or column of a dataframe can be done by using the ‘$’ operator or by using double square brackets with column-name provided within quotes.

```r
student_data$student_ID
student_data[['student_degree']]
student_data['student_degree'] #this returns a dataframe
```

** Accessing Rows of a Data Frame**

To access the rows of the dataframe, we provide the name of the row within single quotes followed by a comma which are together enclosed within square brackets.

```r
student_data[1,] #accesing row1
```

**Accessing Multiple Columns of a Data Frame**

To retrieve values from multiple columns, the name of the column can be provided as a vector.


```r
student_data[c('student_degree','student_marks')]
#accesing multiple columns
```

```r
student_data[c(2,3),] #accesing multiple rows
#note the only difference is the comma after the vector
```

**Factor**

A factor is a vector that can contain only predefined values and is used to store categorical data

A factor can be initialised using the function ‘factor().’ There are built-in functions available that operate on factor objects. For example, the built-in function ‘levels()’ can be used to identify the unique categories inside the vector. Factor objects can also be modified using indexes. A factor data structure will be important when we deal with categorical type data such as gender, education level, blood type, etc.

```r
gender = factor(c('F','M','F','M','F'))
levels(gender)
print(gender)
```
